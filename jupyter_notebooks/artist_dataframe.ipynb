```python
import pandas as pd
import os.path
```

```python
df = pd.read_csv("../datasets/clean_dataset.csv")
```

```python
artist_df = df[["track_id", "track_name", "artists"]].copy()
artist_df["artists"] = artist_df["artists"].str.split(";")
artist_df = artist_df.explode("artists")

# Clean whitespace
artist_df["artists"] = artist_df["artists"].str.strip()
```
