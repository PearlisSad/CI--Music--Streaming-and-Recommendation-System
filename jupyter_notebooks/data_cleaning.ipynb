# Finding Outliers

```python
# Finding and handling outliers
# Simple outlier detection using boxplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

feature = "popularity"

# Boxplot
axes[0].boxplot(df[feature])
axes[0].set_ylabel('Price (Â£)')
axes[0].set_title('Price Distribution - Boxplot')
axes[0].grid(True, alpha=0.3)

# Histogram
axes[1].hist(df[feature], bins=30, edgecolor='black', alpha=0.7)
axes[1].axvline(df[feature].mean(), color='red', linestyle='--', label='Mean')
axes[1].axvline(df[feature].median(), color='green', linestyle='--', label='Median')
axes[1].set_xlabel('Price (Â£)')
axes[1].set_ylabel('Count')
axes[1].set_title('Price Distribution - Histogram')
axes[1].legend()

plt.tight_layout()
plt.show()
```
